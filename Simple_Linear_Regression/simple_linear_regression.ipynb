# Linear Regression
### 1.Data Collection
##### seaborn.tips

```python
import seaborn as sns 
import pandas as pd
df=sns.load_dataset("tips")
df.head()
```

### 2. Uderstanding the dataset

```python
df.info()
```

### 3. stats summary

```python
df.describe()
```

## 4. Relationship check 
#### Before using any model, we must confirm relationship


```python
import matplotlib.pyplot as plt
plt.scatter(df['total_bill'],df['tip'])
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Total Bill vs Tip")
plt.show()
```

# Observations
### 1.As the bill increases tips increases
### 2.Relationship= Straight line (can use linear regression)

## 5.Feature and Target Separation

```python
x=df[['total_bill']] #feature input must be in 2D
y=df['tip'] #feature output can be in 1D
```

## 6.Train - Test Split 
#### (to make sure by testing model, it works well on unseen data also)

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(
    x,y,
    test_size=0.2,
    random_state=42
)
```

##### 80%- training 20%- testing

## 6.Feature Standardization 
#### (features to be in the same scale)
### 1. Compare the coefficients fairly
### 2. Avoid the Dominance of Large values
### 3. Prepare for Ridge/Lasso

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)  #transform only for test data
```

### IMP: 
#### 1.Fit only on training data
#### 2.Prevents the data leakage

## 7.Train Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train_scaled,y_train)
print("coefficient (m):",model.coef_[0])
print("Intercept (c):",model.intercept_)
```

```python
y_pred=model.predict(x_test_scaled)
```

## 8.Model Evaluation

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
print("R2 Score:",r2)
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
mse= mean_squared_error(y_test,y_pred)
rmse= np.sqrt(mse)
print("MSE:",mse)
print("RMSE:",rmse)
```

## 9. Visualization of best fit line

```python
plt.scatter(x,y)
plt.plot(x,model.predict(scaler.transform(x)),color='red')
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Linear Regression Best Fit Line")
plt.show()
```

## 10. Input - Output Real time

```python
bill_amount=float(input("Enter the bill amount:"))
bill_scaled=scaler.transform([[bill_amount]])
predict_tip=model.predict(bill_scaled)
print(f"For a bill amount of {bill_amount},")
print(f"Predicted Tip: {predict_tip[0]:.2f}")
```
